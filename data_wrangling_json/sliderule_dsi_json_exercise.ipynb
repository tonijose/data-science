## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

#### Converting the list of dictionaries of theme names to a new dataframe

```python
newdf = pd.DataFrame([x for x in world_bank_df['mjtheme_namecode']])
newdf
```

#### Merging all columns to a new series

```python
combineddf = pd.DataFrame
combineddf = pd.concat([newdf[0], newdf[1], newdf[2], newdf[3], newdf[4]])
combineddf = pd.DataFrame(combineddf).reset_index()
combineddf.insert(0, 'id', range(1, 1 + len(combineddf)))
combineddf.drop('index',axis = 1, inplace = True)
combineddf.columns = ['index', 'theme']
combineddf
```

#### Conveting the series to a new data frame with columns index, theme code and theme name

```python
theme_df = pd.concat([combineddf['index'], combineddf['theme'].apply(pd.Series)], axis=1)
theme_df
```

#### Replace empty names with Nan

```python
theme_df['name'].replace('', np.nan, inplace=True)
theme_df
```

### Find the top 10 major project themes (using column 'mjtheme_namecode') 

```python
theme_df.groupby(by =['name'])['code'].count().reset_index(name='count').sort_values(['count'], ascending=False).head(10)
```

### In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

#### replace all the missing name column with NA

```python
theme_df['name'].replace('', np.nan, inplace=True)
```

#### forward fill all the missing name columns

```python
#newdf['name'].replace('', np.nan, inplace=True)
theme_df = theme_df.fillna(method='ffill')
```
